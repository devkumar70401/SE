## The McCulloch-Pitts Neuron (1943)

![Screenshot From 2026-04-17 11-57-51.png](<attachment:Screenshot From 2026-04-17 11-57-51.png>)

**The Core Idea: A Simple Biological Logic Gate**

- The first mathematical model of a neuron. It Operates on simple principle; if enough inputs are active, the neuron "fires"

**How it Works**:

- Takes multiple binary inputs $(x_i \in {0,1})$
- Calculates their sum.
- The output is 1 if the sum is greater than or equal to a fixed thresold $(\theta)$
- The output is 0 otherwise.

**Significance**:

- Proved that networks of these simple units could compute any logical function (e.g. AND, OR).

## Critical Limitations of the M-P Neuron

**A calculator, Not a Learner**

- The M-P neuron was a fixed logic gate. It could compute, but it could not learn.

**No Feature Importance**

- Equally important features.
- Can't prioritize key features
- *Missing piece: Weights*

**No Automatic Learning**

- Thresold $(\theta)$ is fixed.
- Must be set manually
- *Missing piece: A Learning Rule*

**Restricted to Binary I/O**

- Inputs must be 0 or 1
- Outputs are only 0 or 1
- Can't handle real-valued data.

## The Leap to Learning: The Perceptron (1958)

![Screenshot From 2026-04-17 11-58-19.png](<attachment:Screenshot From 2026-04-17 11-58-19.png>)

**Innovation 1: Weighted Input**

- Each input is multiplied by a weight.
- Weights represent feature importance.
- The model can now prioritize signals.

**Innovation 2: The Learning Rule**

- Compares prediction $(\hat{y})$ to true label $(y)$
- If wrong, it adjusts the weights to correct the error.

**Significance: The Birth of leanring Machines**

- For the first time, a machine could automatically learn to classify patterns, laying the foundaion for all of modern deep learning.

## Perceptron: Strengths & Limitations

**Strengths**

- **It Learns**: An automatic learning rule adjusts weights from data
- **Feature Importance**: Learns which inputs are more important via weights
- **Convergence Guarentee**: If data is **linearly separable**, it's guarantee to find a solution

**Limitations**

- **The Fatal Flaw: Linear Separability** Only problems where a single straight line (or hyperplane) can separate the classes can be solved
- **Harsh Threshold**: The step function activation is not differentiable, preventing modern gradient-based training.

# Fundamental Tricks in Deep learning

## The Numerization Trick: From Sensation to Numbers

**The Core Problem: Computers don't see, they calculate**

- Humans percieve the world through qualitative sensations.
- Machines only understand quantitative data (**numbers**).

**Our Goal**

- Convert every piece of information about a problem into a list of numbers, called a **vector**.

## The Numerization Trick: Example

**Scenario: A self-Driving Car**

- A car's AI must perceive the road, identify objects, and decide how to act in real-time.
- **Task**: How do we represent its world as Input and Output vectors?

**Input Vectors(What the Car "Sees")**

- **Camera Data**: A massive vector of pixel values (R,G,B for each pixel)
- **LiDAR Data**: A vector of 3D point coordinates (x,y,z) for nearby surfaces.
- **Radar Data**: A vector of object distances and velocities.
- **Car's State**: A vector of current speed, acceleration, GPS location

**Output Vectors (What the Car "Does")**

- **Control Actions**: A vector like '[Steering Angle, Acceleration, Brake]`
- **Object Predictions**: For each detected object, a vector like '[Class, X-Pos, Y-Pos, Velocity]`.

## How we Numerize: Encoding Categories

Objective: How do we convert an abstract category like "Pedestrian" or "Stop Sign" from the car's output vector into numbers?

**Method 1: One-Hot Encoding**

- Represents a class as a binary vector
- One position is "hot"(1), all others are 0.
- **Car Example (Object Class)**:
    - Pedestrian -> [1,0,0]
    - Vehicle -> [0,1,0]
    - Stop Sign -> [0,0,1]
- Simple but inefficient for many classes

**Method 2: Embedding Layers**

- Learns a dense, meaningful vector for each class
- Similar concepts get similar vectors
- **Car Example (Object class)**
    - Stop sign: [0.9,-0.2,...]
    - Pedestrian: [-0.5,0.7,...]
- Efficient & Captures semantic relationships

## Learn the Function Trick

**Data and Functions**

- We Frame problems as data (inputs & outputs).
- we Frame solutions as a **function** that maps inputs to outputs
- The "learning task" is to find this function

**Every Function has two parts**

- **The Form (Architecture)**:
    - The function's template or structure
    - Learning a structured function is more tractable than learning an arbitrary function.
- **The parameters (w)**:
    - The "dials" that tune the function.
    - Learned automatically from data.

E.g., ChatGPT can be seens as one, massive function taking your prompt as input.

## The Parameterization Trick

**Core Idea**: We define our model as a function with a set of learnble parameters (weights and biases), denoted by $(\theta)$. The goal of training is to find the optimal values of $(\theta)$.

**The model as a Function**

$$\hat{y} = f(x;\theta)$$

where:

- x is the numrical data.
- f is the network architecture (e.g. layers, activation functions).
- $\theta$ represents all the weights and biases in the network
- $\hat{y}$ is the model's prediction

## From Learning to Optimization: The Loss Landscape

Parameterization turns "learning" into a clear problem: optimization


**Step-1: Measuring Error**

- A **loss function** scores how "bad" our model is for a given set of parameters $\theta$.
- High loss = Bad model ; Low loss = Good model.

**Step-2: The Goal**

- Our goal is to find the parameters $\theta^*$ that result in the lowest possible loss.

# Feedforward Network Architecture

## Network Topology: Building a Feedforward Network

![Screenshot From 2026-04-17 11-59-26.png](<attachment:Screenshot From 2026-04-17 11-59-26.png>)

1. **The input Layer**
    - Recieves the data and just pass values along.
2. **The Hidden Layer**
    - The network's computational engine.
    - They learn increasingly abstract features
3. **The output Layer**
    - Produces the final result (e.g., a prediction).

- **Feedforward Flow**: Information moves in one direction only: input -> output. No loops.
- **Fully connected**: Every neuron connects to every neuron in the next layer.

## The Standard Forward Model: A Neuron's Calculation

![Screenshot From 2026-04-17 11-59-58.png](<attachment:Screenshot From 2026-04-17 11-59-58.png>)

**Step-1: The linear Part (Weighted Sum)**

- A weighted sum of all inputs is calculated.
- A bias term is added to this sum.
- **Result**: $z=(\Sigma_iw_ix_i)+b$

**Step-: The Non-Linear Part (Activation)**

- The result z is passes through a non-linear activation function g.
- This introduces complexity and allows the network to learn non-linear patterns.

**The Fundamental Building Block**

- This two-step 'Linear Sum -> Non-Linear Activation' process is performed by every single neuron in a deep neural network.

## Linear Model vs ANN

![Screenshot From 2026-04-17 12-00-29.png](<attachment:Screenshot From 2026-04-17 12-00-29.png>)

## The Forward Pass: The Vectorized Layer View

**The Problem & The Solution**

- **Problem**: Calculating neuron by neuron is incredibly inefficient
- **Solution**: Compute an entire layer at once using vectorized operations (i.e., matrix multiplication)

Let's define the terms for a layer l:

- $a^{[l-1]}$: Vector of activations from the previous layer.
- $W^{[l]}$: Weight Matrix, where $W_{ij}$ is the weight from neuron j in the layer l-1 to neuron in layer l
- $b^{[l]}$: Bias vector for the current layer.
- $z^{[l]}$: Vector of weighted sums for the current layer.
- $a^{[l]}$: Vector of activations for the current layer.

**The Two Core Equations (Vectorized)**

- Weighted Sum: $z^{[l]} = W^{[l]}a^{[l-1]} + b^{[l]}$
- Activations: $a^{[l]} = g(z^{[l]})

> This is computationally efficient and perfectly suited for GPUs.

## The Multilayer Perceptron (MLP)

**What is an MLP?**

# Activation Functions

## Why need Activation Functions?

**The Problem: Stacking Linear layers is Useless**

- Without a non-linear activation function, a deep network simply collapses into a single linear model, no matter how many layers it has.

**Without Non-Linearity**:

- A layer is a linear operation: $Wx+b$.
- Stacking them:
  - $L_2(L_1(x)) = W_2(W_1(x) + b_1) + b_2$
- This simplifies to:
  - $(W_2W_1)x + (W_2b_1+b_2)$
- This is just another linear function: $W^{\prime}x+b^{\prime}$

A 100-layer linear network has the same power as a 1-layer network.

**With Non-Linearity(g)**:

- The equation becomes:
  - $g(W_2g(W_1x+b_1) + b_2)$
- This function cannot be simplified
- This allows the network to learn arbitrarily complex, "wiggly" functions

## Sigmoid

**Formula:** 

$$\sigma(z) = \frac{1}{1+e^{-z}}$$

**Properties:**

- Squeezes any real number into the range(0,1).
- Historically popular for its interpretation as a neuron's "firing rate"
- Useful in output layers for binary classification (predicting probabilites)

**Problems:**

- **Vanishing Gradients**: The functino is flat at both ends. The gradient is near zero for large positive or negative inputs, which effectively stops learning in deep networks.
- **Not Zero-Centered:** Outputs are always positive, which can slow down learning.


```python
# Sigmoid
x = torch.linspace(-7,7, 200) # linspace for smooth curve
y = torch.sigmoid(x)             # Direct implementation

plt.plot(x, y, label='tanh')  # plotting

plt.axhline(0, color='black') # adding X-Axis
plt.axvline(0, color='black') # adding Y-Axis
plt.title('Activation Function: Sigmoid')
plt.grid(True, alpha=0.3)
plt.show()
```


## Tanh

**Formula**:

$$tanh(z) = \frac{e^z-e^-z}{e^z+e^-z}$$

**Properties**:

- Squeezes any real number into the range (-1,1).
- **Zero-Centered**: Its major advantage over Sigmoid. This major advantage over Sigmoid. This property helps center the data for the next layer, often speeding up convergence.

**Problems**:

- **Vanishing Gradients:** Like Sigmoid, the function flattens out, and its gradients disappear for large positive or negative inputs, making it difficult to use in very deep networks.

```python
# Tanh
x = torch.linspace(-4,4, 200) # linspace for smooth curve
y = torch.tanh(x)             # Direct implementation

plt.plot(x, y, label='tanh')  # plotting

plt.axhline(0, color='black') # adding X-Axis
plt.axvline(0, color='black') # adding Y-Axis
plt.title('Activation Function: tanh')
plt.grid(True, alpha=0.3)
plt.show()
```

## ReLU

**Formula**:

$$ReLU(z) = max(0,z)$$

**Properties**:

- **Computationally Efficient**: Very fast to compute (just a threshold)
- **No Vanishing Gradient (for z > 0)**: The gradient is a constant 1 for positive inputs, allowing learning signals to propagate deep into the network.
- **Sparsity**: By outputting 0 for negative inputs, it can make the network sparse.

**Problems**:

- **The Dying ReLU**: If a neuron's weights are updated such that its pre-activation z is always negative, it will always output 0. The gradient will also be 0, and the neuron can never recover. It effectively "dies".


```python
# Tanh
x = torch.linspace(-4,4, 200) # linspace for smooth curve
y = torch.relu(x)             # Direct implementation

plt.plot(x, y, label='tanh', linewidth=5)  # plotting

plt.axhline(0, color='black') # adding X-Axis
plt.axvline(0, color='black') # adding Y-Axis
plt.title('Activation Function: tanh')
plt.grid(True, alpha=0.3)
plt.show()
```


## The ReLU Family

- Fixing the "Dying" Problem

**Leaky ReLU**

$$
f(z) = \begin{cases} 
      z & \text{if } z > 0 \\
      \alpha z & \text{if } z \leq 0 
   \end{cases}
$$

(where $\alpha$ is a small constant like 0.01)

- **The Fix:** Introduces a small, non-zero gradient for negative inputs.
- This allows "dead" neurons to be revived. A very common and effective choice.

**GELU (Gaussian Error Linear Unit)**

- A smoother, probabilistic alternative to ReLU.
- Became the standard in state-of-the-art **Transformer** models (e.g. GPT, BERT)

## A Practical Guide to Choosing Activations

**For Hidden Layers**

- **Start with ReLU.** Most common, fastest, and usually works.
- For issues with "dying" neurons, switch to **Leaky ReLU** or **Parameteric ReLU**.
- For Transformer-based models, consider **GELU** or **Swish**.

**For the Output Layer(Task Dependent)**

- Depends on the problem's output range.
- **Binary Classification: Sigmoid** Outputs a probability between 0 and 1.
- **Multiclass Classification: Softmax** Outputs a probability distribution over all classes (all outputs sum to 1).
- **Regression: None(Linear)** The output can be any real number.

**Rule Of Thumb**

- Never use Sigmoid or Tanh in hidden layers of modern deep networks due to vanishing gradient problem. Their use is now almost exclusively in output layers.

# The Forward Pass & Measuring Error

## The Forward Pass

- From input to Prediction
- The forward pass is the process of taking an input vector $x$ and feeding it through the network, layer by layer, to produce a final prediction $\hat{y}$.

The computation for every layer follows the same two-step, vectorized pattern

1. **Linear Combination:** Calculate the weighted sum for all neurons in the layer. $$z^{[l]} = W^{[l]}a^{[l-1]}+b^{[l]}$$
2. **Non-linear Activation:** Apply the activation function element wise. $$a^{[l]} = g(z^{[l]})$$

> The final activation vector $a^{[l]} = g(z^{[l]})$ is the network's prediction $\hat{y}$




## Loss Functions: Quantifying the Error

A loss function $L(y, \hat{y})$ takes the true label $(y)$ and the prediction $(\hat{y})$ and returns a single number representing the error. This error is what we aim to minimize during training.

**For Regression Tasks**: Mean Squared Error **(MSE)**

$$L_{MSE} = \frac{1}{N} \Sigma_{i=1}^N (y_i-\hat{y}_i)^2$$

- Calculated the average squared difference btw true and predicted values.
- Penalizes large errors very heavily.

**For Classification Tasks:** Cross-Entropy Loss

$$L_{CE} = -\Sigma_{c=1}^C y_c \log(\hat{y}_c)$$

- $y_c$ is 1 if c is the true class, 0 otherwise.
- $\hat{y}_c$ is the predicted probability for class c
- Penalizes the model for being confident and wrong.

**Example (3 Classes)**:

- **True:** Cat -> [1,0,0]
- **Good Pred:** [0.8,0.1,0.1] -> Low Loss
- **Bad Pred:** [0.1,0.2,0.7] -> High Loss

# Gradient Descent

## Gradient Descent: The Intuition

**The Analogy: A Hiker on a Foggy Mountain**

- Imagine you are on a mountainside in thick fog and want to get to the lowest valley. What do you do?

You ask two simple questions at every step:

**1. Which way is downhill?**

- You feel the slope under your feet.
- The direction of steepest ascent is the Gradient $(\nabla L)$
- So, you take a step in the **opposite** direction.

**2. How big a step should I take?**

- This is your "Step size" or Learning Rate $(\eta)$.
- Too small: takes forever. Too big: overshoots the valley.

## Why Gradient Descent? Navigating the Loss Landscape

**The Goal**

- Find the parameters $(\theta)$ that minimizes the loss Function $L(y, f(x;\theta))$. This means finding the lowest point in the landscape.

**The Update Rule**

- At each step, we move in the direction of steepest descent:

$$\theta_{new} = \theta_{old}-\eta L(y, f(x;\theta_{old}))$$

**The challenge: Local vs Global Minima**

- **Global Minimum:** The true lowest point; the optimal solution
- **Local Minimum:** A "valley" that traps the algorithm, preventing it from reaching the global minimum.

## GD in Action: Linear Regression

**1. The Model**

- Find the best line $\hat{y} = mx+c$. Parameters are $\theta = {m,c}$

**2. The Loss Function (MSE)**

$$L(m,c) = \frac{1}{N}(y_i-\hat{y}_i)^2$$

**3. The update rules**

$$m := m-\eta\frac{\delta L}{\delta m}$$

$$c := c- \eta \frac{\delta L}{\delta c}$$


## GD : A practical comparison

| Aspect | Batch GD | Stochastic GD (SGD) | Mini-Batch GD |
| :--- | :--- | :--- | :--- |
| **Data Usage** | Entire dataset | Single data point | Mini-batch of data |
| **Update Frequency** | Once per epoch | After each data point | After each mini-batch |
| **Convergence Speed** | Slow | Fast (per iteration) | Moderate |
| **Convergence Stability** | Smooth and stable | Noisy and erratic | Relatively smooth |
| **Computational Efficiency** | Inefficient for large data | Efficient | **Very Efficient (GPU)** |
| **Memory Requirement** | High | Low | Moderate |

**The winner: Mini-Batch Gradient Descent**
For nearly all modern deep learning applications, Mini-Batch GD is the standard. It offers best compromise between the stability of Batch GD and the speed of SGD, while being perfectly suited for parallelization on GPUs.

# Backpropagation
  
- Calculating the Gradients

## Backpropagation's Engine: The Chain Rule

**The Question**

- How does a small change in an early parameter $(w_1)$ affect the final loss $(L)$ at the end of long chain of a computations?

**The Chain Rule provides the answer by multiplying local derivatives.**

Imagine a function $(L(y(x)))$. We want to find how L changes with respect to x, which is $\frac{\delta L}{\delta x}$.

## Scalar Case: A simple Chain

- We know how L changes with respect to y $\frac{dL}{dy}$
- We know how y changes with respect to x  $\frac{dy}{dx}$
- The chain rule lets us link them: 

$$\frac{dL}{dx} = \frac{dL}{dy} \cdot \frac{dy}{dx}$$

**Vector Case (Generalization)**

- For vector functions $y = f(x)$ and $L = g(y)$, the derivates become jacobian matrices, and the chain rule becomes matrix multiplication:

$$\nabla_x L = (J_yL)(J_xy)$$

- This is the mathematical foundation for backpropagation in neural networks.
  

## Visalizing Computations: The Computaional Graph

![Screenshot From 2026-04-17 11-43-02.png](<attachment:Screenshot From 2026-04-17 11-43-02.png>)

**The Two Passes of Automatic Differentiation**

- **Forward Pass:** We start with input values (a,b,c) and compute the values of all nodes from left to right, until we get the final output $(L)$
- **Backpropagation:** We Start from the end with the gradient $\frac{dL}{dL} = 1.$ We then go backwards (right to left), using the chain rule at each node to compute the gradient of L wrt every input of that node.

## Backpropagation by Hand: A Scalar Example

- Let's run backpropagation on $L = (a+b) \cdot c$ with inputs $a=2, b=3, c=6$.

![Screenshot From 2026-04-17 12-21-08.png](<attachment:Screenshot From 2026-04-17 12-21-08.png>)

**Forward Pass (Green)**

- $d = a+b=2+3=5$
- $L = d \times c = 5 \cdot 6 = 30$

**Backward Pass (Red)**

- $\frac{dL}{dL} = 1 \text{ text }$
- $\frac{dL}{dc} =  \frac{dL}{dL} \cdot \frac{dL}{dc} = 1 \cdot d = 5$
- $\frac{dL}{dd} = \frac{dL}{dL} \cdot \frac{dL}{dd} = 1 \cdot c = 6$
- $\frac{dL}{db} = \frac{dL}{dd} \cdot \frac{dd}{db} = 6 \cdot 1 = 6$
- $\frac{dL}{da} = \frac{dL}{dd} \cdot \frac{dd}{da} = 6 \cdot 1 = 6$

![Screenshot From 2026-04-17 13-10-02.png](<attachment:Screenshot From 2026-04-17 13-10-02.png>)

## Backpropagation in a Multilayer Perceptron

- The goal is to compute the gradient of the loss $(L)$ with respect to every weight $W_{ij}^{[l]}$ and bias $b_i^{[l]}$ in the network. We do this by propagating the error signals backwards, layer by layer.
- To update the weights $W^{[l]}$, we need $\frac{\delta L}{\delta W^{[l]}}$.

![Screenshot From 2026-04-17 13-13-16.png](<attachment:Screenshot From 2026-04-17 13-13-16.png>)

**The Three Key Gradient components for a layer l**

- To update the weights $W^{[l]}$, we need $\frac{\delta L}{\delta W^{[l]}}$. Using the chain rule, we decompose this:

![Screenshot From 2026-04-17 13-16-12.png](<attachment:Screenshot From 2026-04-17 13-16-12.png>)

1. **Upstream Gradient:** How the final loss changes with this layer's output. This is passed back from layer l+1.
2. **Activation Gradient:** The derivative of the activation function, $g \prime (z^{[l]})$.
3. **Local Gradient:** How this layer's pre-activation changes with its weights.


## The Weight Update Rules: The Full Derivation

- Let's derive the update rules for the weights $W^{[l]}$ and biases $b^{[l]}$ of a hidden layer l.

**Step-1: Calculate the "Error" Term $\delta^{[l]}$**

- We define an error term $\delta^{[l]} = \frac{\delta L}{\delta z^{[l]}}$, which represents how the loss changes wrt the pre-activation of layer l.

$$\delta^{[l]} = \frac{\delta L}{\delta z^{[l]}} $$
