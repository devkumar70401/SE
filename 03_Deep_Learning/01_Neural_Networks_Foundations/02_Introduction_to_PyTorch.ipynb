
## Introduction To Tensors

- A tensor is a main data structure in PyTorch.
- It is similar to NumPy array, but with two important advantages:
  - tensors can run on GPU
  - tensors support **automatic differentiation** using PyTorch Autograd.

Tensors are used to represent:

- Input data
- model parameters
- intermediate activations
- gradients
- predictions
- losses

Common tensor dimensions:


````python
scalar = torch.tensor(5)              # 0D tensor
vector = torch.tensor([1, 2, 3])      # 1D tensor
matrix = torch.tensor([[1, 2], [3, 4]])  # 2D tensor
````

In deep learning, tensor shape is extremely important.

```
# Image batch
(batch_size, channels, height, width)

# Text batch
(batch_size, sequence_length)

# Tabular data
(batch_size, num_features)
```

### Creating Tensors

#### From Python Lists

````python
x = torch.tensor([1, 2, 3])
y = torch.tensor([[1, 2], [3, 4]])
print(f"Shape of x: {x.shape}")  # Output: torch.Size([3])
print(f"Shape of y: {y.shape}")  # Output: torch.Size([2, 2])
````

#### Zeros and Ones

````python
zeros = torch.zeros(3, 4)
ones = torch.ones(2, 3)
print(f"Zeros tensor:\n{zeros}")
print(f"Ones tensor:\n{ones}")
````

#### Random Tensors

- `torch.randn()` is commonly used for initializing weights.

````python
rand = torch.rand(3, 4)      # values between 0 and 1
randn = torch.randn(3, 4)    # values from normal distribution
randint = torch.randint(0, 10, (3, 4))  # random integers between 0 and 9
print(f"Random tensor (0 to 1):\n{rand}")
print(f"Random tensor (normal distribution):\n{randn}")
print(f"Random integer tensor (0 to 9):\n{randint}")
````

#### Range Tensors

````python
x = torch.arange(0, 10)
y = torch.arange(0, 10, 2)
z = torch.linspace(0, 1, steps=5)
print(f"Arange (0 to 9):\n{x}")
print(f"Arange (0 to 8, step 2):\n{y}")
print(f"Linspace (0 to 1, 5 steps):\n{z}")
````

#### Creating Tensors Like Another Tensor

````python
x = torch.ones(3, 4)

zeros_like = torch.zeros_like(x)
ones_like = torch.ones_like(x)
rand_like = torch.rand_like(x)
print(f"Zeros like x:\n{zeros_like}")
print(f"Ones like x:\n{ones_like}")
print(f"Random like x:\n{rand_like}")
````

### Reshaping tensors

Reshaping changes the shape of a tensor without changing its data.

#### `reshape()`


````python
x = torch.arange(12)
print(f"Original tensor x:\n{x}")
x = x.reshape(3, 4)
print(f"Reshaped tensor x:\n{x}")
````

#### `view()`

- view() is faster in some cases, but requires tensor to be contiguous in memory.
- `reshape()` is safer because it works even when the tensor is not contiguous.

````python
x = torch.arange(12)
print(f"Original tensor x:\n{x}")
x = x.view(3, 4)
print(f"Viewed tensor x:\n{x}")
````

#### Flattening

````python
x = torch.randn(2, 3, 4)
x_flat = x.reshape(-1)
print(f"Shape of original x: {x.shape}")
print(f"Shape of flattened x: {x_flat.shape}")
````

````python
x = torch.randn(32, 3, 28, 28)
print(f"Shape of x: {x.shape}")
x = x.reshape(32, -1)
print(f"Shape of x after flattening: {x.shape}")
````

The -1 means PyTorch automatically calculates that dimension.

#### Adding Dimension

````python
x = torch.tensor([1, 2, 3])
print(f"Shape of x: {x.shape}")
x = x.unsqueeze(0)
print(f"Shape of x after unsqueeze: {x.shape}")
````

Useful for adding a batch dimension.

#### Removing Dimensions

- Removes dimensions of size 1.

````python
x = torch.randn(1, 3, 1, 5)
print(f"Original shape of x: {x.shape}")
x = x.squeeze()
print(f"Shape of x after squeeze: {x.shape}")
````

> Be Careful
>
> - `x.squeeze()` can accidently remove the batch dimension if batch size is 1.

#### Rearranging dimensions

````python
x = torch.randn(32, 224, 224, 3)
x = x.permute(0, 3, 1, 2)
print(f"Shape of x after permute: {x.shape}")
````

## Tensor Operations

- Tensors operations are the core of deep learning in PyTorch.

### Arithmetic Operations

- Arithmetic operations are basic mathematical operations performed on tensors.
- Common arithmetic operations:

```py
A + B   # addition
A - B   # subtraction
A * B   # multiplication
A / B   # division
A ** 2  # power
```

````python
import torch

a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

print(a + b)  # tensor([5, 7, 9])
print(a * b)  # tensor([4, 10, 18])
````

Arithmetic operations can be performed:

- Tensor with tensor
- Tensor with scalar
- Tensor with broadcastable tensor

````python
a = torch.tensor([1, 2, 3])
print(a + 10)  # tensor([11, 12, 13])
````

### Broadcasting

- Broadcasting allows operations between tensors of different shapes.
- Here, b is automatically expanded to match the shape of a.

Broadcasting is important because it avoids writing loops and makes computation faster.

Broadcasting rules:

- Dimensions are compared from right to left.
- Dimensions must either be equal, or one of them must be 1.
- If shapes are incompatible, an error occurs.

```py
(2, 3) and (3,)  -> valid
(4, 1) and (4, 5) -> valid
(2, 3) and (2,) -> invalid
```

````python
a = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

b = torch.tensor([10, 20, 30])

print(a + b)
````

### Matrix Multiplication

- Matrix multiplication is one of the most important tensor operations in neural networks.
- It is different from element-wise multiplication.
- Matrix multiplication shape rule:
  - The inner dimensions must match.
```py
A shape = (m, n)
B shape = (n, p)

A @ B shape = (m, p)
```

Matrix multiplication is used in:

```
Dense layers
Attention mechanisms
Embedding projections
Convolution internally
Transformer models
Backpropagation
```

````python
# Element-Wise Multiplication
A  = torch.tensor([[1, 2],
                   [3, 4]])
B = torch.tensor([[5, 6],
                   [7, 8]])

A * B
````

````python
# Matrix Multiplication
A @ B # OR torch.matmul(A, B)
````

In neural networks, matrix multiplication is used for fully connected layers:

```py
output = input @ weights + bias
```

if
```py
input shape   = (batch_size, input_features)
weights shape = (input_features, output_features)
bias shape    = (output_features)
```

then
```py
output shape = (batch_size, output_features)
```

#### Dot Product

- The dot product is a special case of matrix multiplication between two vectors.
- In deep learning, dot products are used to compute similarity and weighted sums.



````python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

print(torch.dot(a, b))
````

### Aggregation Operations

- Aggregation operations reduce multiple tensor values into fewer values.

Common aggregation operations:

```py
torch.sum()
torch.mean()
torch.max()
torch.min()
torch.argmax()
torch.argmin()
torch.prod()
torch.std()
torch.var()
```

````python
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(torch.sum(x))   # tensor(21)
print(torch.mean(x.float()))  # tensor(3.5000)
print(torch.max(x))   # tensor(6)
print(torch.min(x))   # tensor(1)
print(torch.std(x.float()))  # tensor(1.8708)
print(torch.argmax(x))  # tensor(5) - index of max value in flattened tensor
print(torch.argmin(x))  # tensor(0) - index of min value in flattened tensor
print(torch.unique(x))  # tensor([1, 2, 3, 4, 5, 6])
print(torch.nonzero(x > 3))  # tensor([[1, 0], [1, 1], [1, 2]]) - indices of elements greater than 3
print(torch.where(x > 3, x, torch.tensor(0)))  # tensor([[0, 0, 0], [4, 5, 6]]) - replace elements > 3 with their value, else 0
print(torch.any(x > 5))  # tensor(True) - checks if any element is greater than 5
print(torch.all(x > 0))  # tensor(True) - checks if all elements are
print(torch.all(x > 3))  # tensor(False) - checks if all elements are greater than 3
````

Aggregation can be done over the entire tensor or along a specific dimension.

````python
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print(torch.sum(x, dim=0))
print(torch.sum(x, dim=1))
````

In deep learning, aggregation operations are used in:

```
Loss functions
Accuracy calculation
Pooling layers
Batch statistics
Gradient reduction
Attention score normalization
Global average pooling
```

### Element-wise Operations

Element-wise operations apply the same operation to each element of a tensor independently.

````python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

print(a * b)
print(a.mul(b))  # element-wise multiplication
print(torch.mul(a, b))  # element-wise multiplication
````

Common element-wise operations:

```py

torch.mul()

```

Element-wise operations are heavily used in activation functions.


### In-place Operations

- In-place operations modify the original tensor directly instead of creating a new tensor.
- In PyTorch, in-place operations usually end with an underscore _.


````python
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

x = x.to(torch.float32)

print(x.add_(5))
print(x.sub_(2))
print(x.mul_(3))
print(x.div_(2))
print(x.relu_())
````

````python
x = torch.tensor([1, 2, 3])
x.add_(10)

print(x)
````

- In-place operations save memory because they do not create a new tensor.
- However, they can be dangerous in deep learning because they may interfere with automatic differentiation.
- PyTorch tracks operations for backpropagation. If a tensor needed for gradient calculation is modified in-place, PyTorch may raise an error or compute incorrect gradients.
- Important exam point:
  - In-place operations are memory efficient but should be used carefully when gradients are involved.

### Common Device Mismatch errors

- A device mismatch error happens when tensors involved in the same operation are on different devices.

Common Causes of device mismatch:

1. Model is on GPU but input is on CPU.
2. Labels are left on CPU while outputs are on GPU.
3. New tensors are created inside training loop without specifying device.
4. Hidden states in RNNs are created on CPU while model is on GPU.
5. NumPy arrays are converted to tensors but remain on CPU.


## Reproduccibility

Reproducibility means getting the same or nearly same results when running the same code again.

In deep learning, reproducibility is important for:

- Debugging
- Comparing models
- Scientific experiments
- Examining whether improvements are real
- Sharing results

Deep learning results can vary because many operations involve randomness.

### Random seeds

A random seed fixes the starting point of a random number generator.

If the same seed is used, the same sequence of random numbers is generated.

Example:


````python
import torch

torch.manual_seed(42)

x = torch.rand(3)
print(x)
````

Set seeds for PyTorch:

````python
torch.manual_seed(42)
````

For CUDA:

````python
torch.cuda.manual_seed(42)
torch.cuda.manual_seed_all(42)
````

For Python random module:

````python
import random
random.seed(42)
````

For NumPy:

````python
import numpy as np
np.random.seed(42)
````

### Deterministic Behavior

Deterministic behavior means the same input and same code produce the same output every time.

In PyTorch, some GPU operations may be non-deterministic because they use optimized parallel algorithms.

To request deterministic behavior:

````python
torch.use_deterministic_algorithms(True)
````

Important exam point:

> Deterministic settings can make training slower because PyTorch may avoid faster non-deterministic algorithms.

### Why Results May Still Differ

Even after setting seeds, results may still differ slightly.

Reasons include:

1. GPU non-determinism
   - Some CUDA operations use parallel execution where operation order may vary.
2. Floating-point precision
   - Floating-point arithmetic is approximate.

Example:

```
(a + b) + c may not equal a + (b + c)
```

3. Different hardware
    - CPU, GPU, and different GPU models may produce slightly different results.
4. Different PyTorch/CUDA/cuDNN versions
    - Different library versions may use different algorithms.
5. Non-deterministic operations
   - Some operations do not have deterministic implementations on all devices.
6. Multi-threading
   - Parallel CPU/GPU operations may execute in different orders.
7. DataLoader workers
    - Multiple workers can introduce randomness if not seeded properly.

# Automatic Differentiation

Automatic differentiation is the technique deep learning libraries use to automatically calculate gradients.

In neural networks, we need gradients to know:
> How should each weight change so that the loss becomes smaller?


Instead of manually differentiating every equation, PyTorch uses autograd to track tensor operations and calculate derivatives automatically.

## What is Autograd?

### Autograd

- Autograd means automatic gradient calculation.
- PyTorch’s autograd system records operations performed on tensors and uses them to compute gradients during backpropagation.

````python
import torch

x = torch.tensor(2.0, requires_grad=True)

y = x ** 2

y.backward()

print(x.grad)
````

Explanation

```
y = x²
dy/dx = 2x
when x = 2, dy/dx = 4
```

PyTorch automatically calculated the derivative.

### Computational Graphs

- A computational graph is a graph that represents mathematical operations.
- It shows how inputs are transformed into outputs.

Example:

````python
x = torch.tensor(2.0, requires_grad=True)
y = x * 3
z = y + 4
loss = z ** 2
````

Mathematically:

```
y = 3x
z = y + 4
loss = z²
```

Graph:

```
x → multiply by 3 → y → add 4 → z → square → loss
```

Each operation becomes a node in the graph.

The graph stores information needed to calculate gradients later.

Important terms:

|Term|Meaning|
|---|---|
|Node|Operation or tensor in the graph|
|Edge|Connection showing dependency|
|Input tensor|Starting tensor|
|Output tensor|Result tensor|
|Loss|Final scalar value usually used for backpropagation|

In PyTorch, computational graphs are usually dynamic.

This means the graph is created while the code runs.

This is called:
```
define-by-run
```
