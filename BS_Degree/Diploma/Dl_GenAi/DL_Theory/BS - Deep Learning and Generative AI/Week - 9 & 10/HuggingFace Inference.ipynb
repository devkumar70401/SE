## 🤗 What is Hugging Face?

Hugging Face is an open-source platform and AI community that provides tools, models, and datasets for **Natural Language Processing (NLP)**, **Computer Vision**, **Audio**, and **Multimodal AI** tasks.  
It makes **state-of-the-art AI accessible to everyone** — from beginners to research professionals — with minimal setup and zero retraining.

### 🧩 Key Highlights

1. **Model Hub**  
   - A central repository hosting **thousands of pretrained models** for NLP, vision, and audio.  
   - Supports architectures like BERT, GPT, T5, RoBERTa, and Vision Transformers.

2. **The `transformers` Library**  
   - A popular Python library that allows easy access to pretrained models.  
   - Provides ready-to-use **pipelines** for tasks such as text classification, translation, question answering, summarization, and more.  
  

3. **Datasets Library**  
   - Offers a wide collection of **standardized datasets** for various ML tasks.  
   - Efficiently handles loading, preprocessing, and streaming large datasets.

4. **Tokenizers Library**  
   - Provides fast, customizable tokenization optimized in Rust and accessible from Python.  
   - Handles large text corpora efficiently for model preprocessing.

5. **Inference API & Spaces**  
   - **Inference API:** Allows easy model deployment and real-time predictions.  
   - **Spaces:** Lets you build and host interactive ML apps using **Gradio** or **Streamlit** — free for open-source projects.

## ⚙️ What is `transformers.pipeline`?

The `pipeline` is a **high-level API** provided by the Hugging Face `transformers` library that allows you to perform **end-to-end NLP and vision tasks** using pretrained models — with just a few lines of code.

It basic terms, `pipeline` takes care of model loading, tokenization, and inference, making it ideal for beginners, prototypes, and quick experimentation.

---

### 🧩 Key Features

1. **Ease of Use**  
   - Simplifies complex model workflows into a single callable object.  
   - No need to manually handle tokenizers or model architectures.  


2. **Supports Multiple Tasks**  
   The `pipeline` supports a wide range of tasks out of the box, including:
   - 🗣️ **Text Classification** (Sentiment Analysis, Topic Detection)  
   - 🧠 **Named Entity Recognition (NER)**  
   - ❓ **Question Answering**  
   - 📜 **Text Summarization**  
   - 🌍 **Translation**  
   - 🧩 **Feature Extraction**  
   - ⚖️ **Zero-Shot Classification**  
   - 🖼️ **Image Classification** and **Object Detection** (for vision models)

# Sentiment Analysis

## 💬 What is Sentiment Analysis?

**Sentiment Analysis** is an NLP task that determines the **emotional tone or opinion** expressed in a piece of text — whether it is **positive**, **negative**, or **neutral**.

It is commonly used in **product reviews**, **social media monitoring**, and **customer feedback analysis** to understand public sentiment.

---

### 📘 Example

**Input:**  
`"I absolutely love this phone! The camera quality is amazing."`

**Output:**  
**Sentiment:** Positive ✅

---

### 🛠️ Objective
Now lets try inferencing a few State-Of-The-Art models from HuggingFace for the Sentiment Analysis, without any further training.


```python
# To avoid printing of any warning messages:

import warnings
warnings.filterwarnings("ignore")
```

```python
# import the transformers.pipeline :
from transformers import pipeline
print("Loading default sentiment analysis pipeline...")


# This is the simple way of loading a task specific model
sentiment_classifier = pipeline("sentiment-analysis")

# This will typically load 'distilbert-base-uncased-finetuned-sst-2-english' model
# which is trained on movie reviews (SST-2 dataset) and has 2 labels: POSITIVE, NEGATIVE

```

```python
#sample input:
texts = ["This was a pathetic film.",
         "It was a wonderful acting by the hero, that made the film look good."]

print(f"\n--- Simple Pipeline (Default Model) ---")
print()

for text in texts:
  # Forward pass the input through the model pipeline and extract final output
  result = sentiment_classifier(text)
  print(f"Text: {text}")
  print(f"Result: {result}")
  print()
```

---

Now lets see how to use a specific model for the same Sentiment Analysis task.
---

---

```python
print("Loading specific sentiment analysis pipeline (Twitter-RoBERTa)...")

# We specify a model trained on tweets, which has 3 labels (Positive(LABEL_2), Neutral(LABEL_1), Negative(LABEL_0))
model_name = "cardiffnlp/twitter-roberta-base-sentiment"

sentiment_classifier_specific = pipeline("sentiment-analysis", model=model_name)
```

```python
#sample input:
texts = ["I absolutely love this phone. The camera quality is amazing.",
         "He is a middle class person.",
         "Mr.X is a very pathetic human being"]


print(f"\n--- Specific Model (cardiffnlp/twitter-roberta-base-sentiment) ---")
print()

for text in texts:
  result = sentiment_classifier_specific(text)
  print(f"Text: {text}")
  print(f"Result: {result}")
  print()
```



# Named Entity Recognition(NER)

## What is Named Entity Recognition?
**Named Entity Recognition (NER)** is an NLP task that identifies and classifies key elements in text into predefined categories such as **person names**, **organizations**, **locations**, **dates**, and more.

It helps extract **structured information** from unstructured text — useful in information retrieval, question answering, and knowledge graph building.

---

### 📘 Example

**Input:**  
`"Barack Obama was born in Hawaii and served as President of the United States."`

**Output:**  
- **Barack Obama** → PERSON  
- **Hawaii** → LOCATION  
- **United States** → LOCATION  

---

### 🛠️ Objective
Now lets try inferencing a few State-Of-The-Art models from HuggingFace for the Named Entity Recognition, without any further training.

```python
print("Loading default NER pipeline...")

# The default model is 'dbmdz/bert-large-cased-finetuned-conll03-english'
ner_pipeline = pipeline("ner", grouped_entities=True)     # `grouped_entities` parameter is enabled so that each token representing same tags be merged in output, for better displaying.
```

```python
#Sample Inputs:
texts = ["My name is S Ramanujam, and I am from Chennai.",
         "Barack Obama was born in Hawaii and served as President of the United States."]

print(f"\n--- Simple Pipeline (Default Model) ---")
print()

for text in texts:
  entities = ner_pipeline(text)
  print(f"Text: {text}")
  print("Found Entities:")
  for entity in entities:
      print(f"  - Entity: {entity['word']}, Label: {entity['entity_group']}, Score: {entity['score']:.4f}")
  print()
```

---

Now lets see how to use a specific model for the same NER task.
---

---

```python
print("Loading specific NER pipeline (BERT Large CoNLL-03)...")


model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
ner_pipeline_specific = pipeline("ner", model=model_name, grouped_entities=True)

```

```python
#Sample Inputs:
texts = ["My name is S Ramanujam, and I am from Chennai.",
         "Barack Obama was born in Hawaii and served as President of the United States."]

print(f"\n--- Specific Model {model_name} ---\n")

for text in texts:
  entities = ner_pipeline_specific(text)
  print(f"Text: {text}")
  print("Found Entities:")
  print(entities)
  for entity in entities:
      print(f"  - Entity: {entity['word']}, Label: {entity['entity_group']}, Score: {entity['score']:.4f}")
  print()
```

# Machine Translation

## 🌍 What is Machine Translation?

**Machine Translation (MT)** is an NLP task that automatically translates text from one language to another using pretrained models.  
It helps bridge language barriers and enables global communication in real time.

---

### 📘 Example

**Input (English):**  
`"The quick brown fox jumps over the lazy dog."`

**Output (French):**  
`"Le renard brun rapide saute par-dessus le chien paresseux."`

---

### 🛠️ Objective
Now lets try inferencing a few State-Of-The-Art models from HuggingFace for the Named Entity Recognition, without any further training.



```python
print("Loading default EN-to-FR translation pipeline...")

# This implies a default model, 'google-t5/t5-base'
translator_default = pipeline("translation_en_to_fr")
```

```python
text = "The quick brown fox jumps over the lazy dog."
translation = translator_default(text)

print(f"\n--- Simple Pipeline (EN-to-FR Default) ---")
print(f"Original (EN): {text}")
print(f"Translation (FR): {translation[0]['translation_text']}")
```

---

Now lets see how to use a specific model for the same Machine Translation task.
---

---

```python
print("Loading specific EN-to-DE translation pipeline...")

# Here we specify the task is "translation" and provide the exact model we want
# We perform translation from english to germany
model_name = "Helsinki-NLP/opus-mt-en-de"
translator_specific = pipeline("translation", model=model_name)

```

```python
text = "The quick brown fox jumps over the lazy dog."
translation = translator_specific(text)

print(f"\n--- Specific Model (Helsinki-NLP/opus-mt-en-de) ---")
print(f"Original (EN): {text}")
print(f"Translation (DE): {translation[0]['translation_text']}")
```

# Using the Gemini API

## 📝 Note

- **Gemini** is Google’s **state-of-the-art conversational foundation model**, available in multiple variants.  
- Unlike the open-source models we explored earlier, **Gemini is a closed-source model** — its weights are proprietary to Google and **not publicly released**.  
- Therefore, **Gemini and its variants are not hosted** on Hugging Face or any other open-source platforms.  
- The **only way to access Gemini** is through the **official Gemini API** provided by Google.

---

### 🛠️ Objective

In this section, we’ll explore how to use the **free Gemini API** offered by **Google AI Studio** to perform a few **NLP inference tasks** such as text generation, summarization, and question answering.


## ⚠️ Important Note

The API used in this notebook is part of the **free-tier access** provided by **Google AI Studio**, and it comes with a **predefined token limit**.  

Please **do not use it for commercial purposes** or in any **production or deployable applications**.  For such use cases, refer to the **paid plans and usage policies** of the official **Gemini API** offered by Google.


## Setting up the API:

1. Open your Google Colab notebook and click on the 🔑 Secrets tab in the left panel. <img src="https://storage.googleapis.com/generativeai-downloads/images/secrets.jpg" alt="The Secrets tab is found on the left panel." width=50%>

```python
import google.generativeai as genai
```

```python
from google.colab import userdata
key = userdata.get('GEMINI_API_KEY')
```

```python
# Configure your API key (replace with your actual key or use environment variable)
genai.configure(api_key=key)

# Initialize the `gemini-2.5 pro` model
model_name = "gemini-2.5-pro"
model = genai.GenerativeModel(model_name)   #we can also use other variants of gemini like `gemini-2.5-flash`, `gemini-2.5-flash-lite`, `gemini-2.0-flash`, `gemini-2.0-flash-lite`, etc
print(f"Successfully initialized {model_name} model")
```

```python
#1. Text Generation Task
response = model.generate_content("Prove the Pythagoras theorem of triangles")

# The model.generate_content() method sends the input prompt to the Gemini model
# and returns a structured response object that includes the model's generated text.
# This object can also include metadata, safety attributes, or candidate responses
# depending on the Gemini API version used.

print(response.text)
```

```python
#2. Machine Translation Task
response = model.generate_content("Translate this into hindi: 'The cat ran away'")
print(response.text)
```

```python
# 3. Text Summarization Task
response = model.generate_content('''Summarize the following text in 2-3 lines:
"The pipelines are a great and easy way to use models for inference. These pipelines are objects that abstract most of the complex code from the library, offering a simple API dedicated to several tasks, including Named Entity Recognition, Masked Language Modeling, Sentiment Analysis, Feature Extraction and Question Answering. See the task summary for examples of use.

There are two categories of pipeline abstractions to be aware about:

    The pipeline() which is the most powerful object encapsulating all other pipelines.
    Task-specific pipelines are available for audio, computer vision, natural language processing, and multimodal tasks.

The pipeline abstraction

The pipeline abstraction is a wrapper around all the other available pipelines. It is instantiated as any other pipeline but can provide additional quality of life."''')
print(response.text)
```
