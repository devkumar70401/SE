## 1.2.The XOR Problem
- XOR is the simplest problem a single layer neural network cannot solve.
- It demonstrates exactly why we need hidden layers and non-linear activation functions.
- Single layer neural network = Perceptron
- XOR returns 1 only if inputs are different and 0 if inputs are same
- `AND` and `OR` are linearly separable

NOTE:
-  In coding, Neural Networks are simply classes.
- In `nn.Linear(in_features, out_features)`:
  - in_features or first parameter defines the number of input coming from previour layer or number of neurons of previous layer.
  - out_features or second parameter defines the number of ouputs form this current layer or number of neurons in current layer.
- In input layer, the first argument `in_features` is the number of features per sample not the number of samples.
- We don't have to use activation function for input layer(i.e. first layer of NN)
- We have to use activation function necessarily for hidden layers
- whether to use activation function on output layer is task dependent.

Mental Model
- Input: The raw data (not a layer in the code).
- Hidden Layer(s): Defined using `nn.Linear` (where `in_features` matches the input data dimension).
- Output Layer: The final `nn.Linear` layer that produces your prediction

```python
# Data and Labels
import torch
import torch.nn as nn

# Inputs
X = torch.tensor([[0,0], [0,1], [1,0], [1,1]], dtype=torch.float32)

# Outputs
y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)
```

```python
class XORModel(nn.Module):
  def __init__(self):
    super().__init__()

    # Layer 1 (First Hidden Layer): Takes 2 inputs, outputs 2 hidden features
    self.layer1 = nn.Linear(in_features=2, out_features=2)


    # Layer 2 (Output Layer): Takes 2 hidden features, outputs 1 prediction
    self.layer2 = nn.Linear(in_features=2, out_features=1)

    # Activation
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    x = self.sigmoid(self.layer1(x))
    x = self.sigmoid(self.layer2(x)) # Final output btw 0 and 1
    return x

model = XORModel()
```

Loss and Optimization

- for standard 0/1 problems we use BCE loss.
- for optimisation we can use SGD.


NOTE: `nn.Linear()` automatically creates weight matrix and bias vector. we do not have to manually define them.

```python
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.1)
```

The Training Loop
1. Forward Pass
    - First, input data(X) flows through the network, layer by layer, until it produces an output prediction (`y_pred`)
    - At each neuron and layer, intermediate values(activation) are calculated and often stored. These are crucial for the backward pass.
2. Calculate Loss
    - The `y_pred` is compared to the true labels (`y`) using a chosen loss function. This gives us a single scalar value representing the overall error of the network for the current input.
3. Zero Gradients
    - To forbid acumulating gradients.
4. Backward Pass (Backpropagation)
    - Backpropagation starts at the output layer. We calculate the gradient of the loss with respect to the output of the last layer's activation function. This tells us how much the output layer's activation needs to change to reduce the loss.
    - Then, using the chain rule of calculus, we compute the gradients of the loss with respect to the weights and biases of the output layer. These are often denoted as `dl/dW_output` and `dl/db_output`.
    - Now these gradients are propagated backward through the network, layer by layer, using the chain rule.
    - For each layer we calculate two main things:
      - **Gradients wrt the inputs of the current layer:** These gradients become the "error signal" for the previous layer. They tell the previous layer how its outputs should change to contribute less to the overall error.
      - **Gradients wrt the weights and biases of the current layer:** These are the gradients that we ultimately need for updating the parameters of this specific layer.
    - This process continues until we reach the input layer, calculating the gradients for all weights and biases in all hidden layers.
5. Optimization (Updation)
    - Once all gradients (`dl/dW` and `dl/db` for every parameter) have been calculated for the entire network, the optimier takes over.
    - It uses these gradients to update the weights and biases of the network.
    - A common update rule, like in SGD is `new_weight = old_weight - learning_rate * dl/dW` `new_bias = old_bias - learning_rate * dl/db`
    - The `learning_rate` is a hyperparameter that controls the step size of these updates.

```python
epochs = 20000 # number of times to repeat

for epoch in range(epochs):
  # 1. Forward Pass
  y_pred = model(X)

  # 2. Calculate Loss
  loss = loss_fn(y_pred, y)

  # 3. Zero Grad
  optimizer.zero_grad()

  # 4. Backward Pass
  loss.backward()

  # 5. Updation
  optimizer.step()

  if epoch%4000 == 0:
    print(f"Epoch {epoch}: Loss {loss.item():.4f}")
```

- `optimizer.zero_grad()`
    - `zero_grad()` is a method of the optimizer object
    - Its purpose is to clear gradients of all the parameters(weights and biases) that the optimizer is tracking.
    - When we pass `model.parameters()` in optimizer from then onwards optimizers linked to our model.
- `loss.backward()`
    - loss is a scalar  tensor.
    - Every tensor in pytorch has backward method.
    - `loss.backward()` triggers the pytorch autograd engine to perform the backward pass (backpropagation) starting from the loss tensor, it traverses the computational graph backward.
- `optimizer.step()`
    - `step()` is a method of optimizer object.
    - it uses gradients to update the parameters.

```python
with torch.no_grad():
  final_prediction = model(X)
  print(f"\nFinal Prediction (rounded):\n {final_prediction.round()}")
```

Key Takeaway
- If loss is still relatively large and appears to be consistently decreasing during training, then we should increase number of epochs.

```python
# Truth Table

X = torch.tensor([[0,0], [0,1], [1,0], [1,1]], dtype=torch.float32)
y = torch.tensor([[0], [1], [1], [0]], dtype=torch.float32)
```

```python
# Imports
import torch
import torch.nn as nn

# Set Manual seed
torch.manual_seed(42)

# Model
model = nn.Sequential(
    nn.Linear(2,2), # 1st Hidden layer
    nn.Sigmoid(),  # Activation function for hidden layer
    nn.Linear(2,1), # Output layer
    nn.Sigmoid()   # Activation function for output layer
)

# Loss Function and Optimizer
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.5)

# Training loop
EPOCHS = 20000
for epoch in range(EPOCHS):
  y_pred = model(X)
  loss = loss_fn(y_pred, y)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch % 4000==0:
    print(f"Epoch {epoch}: loss: {loss.item():.4f}")

# Evaluation
with torch.inference_mode():
  final_pred = model(X)
  print(f"\nFinal Prediction:\n {final_pred.round()}")
```

### 1.3.1.Data Loading

- `ToTensor()` converts pixels from [0,255] to [0,1] and reorders HWC -> CHW.
- `Normalize()` centers data around 0 for stable training.
    - In normalize first component is mean and second is standard deviation
- `transforms.Compose()` chains multiple steps into one pipeline

The Parameters
- `root`: it decides where the dataset will be downloaded, if the folder doesn't exists it will create one.
- `train=True/False`: MNIST dataset is pre-split by the creators.

Why Transform before download?
- We aren't actually applying the transformation to the files on the internet. We are attaching a set of instructions to the dataset object.
- This is called "On-the-fly" processing. It's better because it doesn't change the raw files on your disk, and it saves RAM because you only transform the small "batch" you are currently using.

Understanding train data better.
- `train_data`: This is FashionMNIST dataset object.
- `train_data[0]`: This retrieves the first sample from the training dataset. Each sample in FashionMNIST is a tuple containing `(image_tensor, label_integer)`.

```python
# Defines transforms
transform = transforms.Compose([
    transforms.ToTensor(),          # [0,255] -> [0,1] and HWC -> CHW
    transforms.Normalize((0.5,), (0.5,))          # center to [-1,1]
])

# Download and load datasets
train_data = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)
test_data = datasets.FashionMNIST(root='./data', train=False, download=True, transform=transform)

# Data Loaders :  Handles batching and shuffling
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)

# Class names for visualization
class_names = ['T-shirt/Top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal',
               'Shirt', 'Sneaker', 'Bag', 'Ankle Boot']


# Key info
print(f"Training Samples: {len(train_data):,}")  # : ->format specifier & , -> thousands separator
print(f"Test Samples: {len(test_data):,}")
print(f"Image Shape: {train_data[0][0].shape}") # CHW
print(f"Number of Classes: {len(class_names)}")
print(f"Batches per epoch: {len(train_loader)}")
```

### 1.3.3.The Model Architecture
- An MLP that flattens 28 x 28 images to 784 inputs, passes through two hidden layers and ReLU and dropout, and outputs 10 raw logits. no softmax in forward() - crossentropyloss handles it.

Architecture: 784 -> 128 -> 64 -> 10

1. The Numbers: `784 -> 128 -> 64 -> 10`
    - These represent the Neurons in each layer.
    - 784 (Input): This is fixed. $28 \times 28$ pixels = 784 numbers.
    - 128 & 64 (Hidden Layers): These numbers are Hyperparameters. There is no "law" that says it must be 128. It could be 200, 50, or 1000. We choose powers of 2 (like 64, 128, 256) because computers (GPUs) are slightly faster at processing these sizes.
2. ReLU vs Sigmoid
    - Sigmoid (what you used in XOR) is old-fashioned. It has a problem: when numbers get very large or small, Sigmoid "flattens out," making it very slow for the model to learn (this is called the Vanishing Gradient problem).
    - ReLU (Rectified Linear Unit) is the "New Standard." Its math is simple:
        - If the number is negative, it becomes 0.
        - If the number is positive, it stays the same. This simplicity makes deep networks train much, much faster.
3. Dropout (The Strict Teacher)
    - Dropout is a clever trick to prevent Overfitting (when a model just memorizes the data instead of learning patterns).
        - During training, Dropout randomly "turns off" some neurons (usually 20% or 50%) in every batch.
        - This forces the remaining neurons to work harder and not rely on "neighbor" neurons. It’s like a teacher telling students they can't sit next to their best friends during a test.
4. Raw Logits vs softmax
    - Logits: These are the raw, unscaled scores coming out of the last layer. They can be any number (e.g., $-5.2$, $10.4$, $0.1$).
    - Softmax: A function that turns those raw scores into probabilities that add up to 100% (e.g., $0.01$, $0.98$, $0.01$).
    - Why skip softmax in the code?
        - Your instructor is using `nn.CrossEntropyLoss()`. In PyTorch, this specific loss function has Softmax "built-in" to the math. If you add Softmax yourself and then use CrossEntropyLoss, you are doing it twice, which can actually hurt the math!


### 1.3.4.**Dropout: The concept**
1. The Core Problem: Co-dependency
    - In a normal neural network, neurons often become "lazy" and start relying on a few "superstar" neurons to do all the work. This is called Co-dependency.
    - The Risk: The model stops learning the actual features of a "Shirt" and just memorizes specific pixel patterns that the superstar neurons picked up. This leads to Overfitting (the model is great at the training data but fails in the real world).
2. How Dropout Works (The "Strict Teacher")
    - When you apply a Dropout of 0.5 (50%):
        - During Training: In every single "batch" of images, PyTorch randomly picks 50% of the neurons and sets their values to zero.
        - The Result: The "superstar" neurons are gone! Now, the other neurons are forced to learn how to identify the image on their own.
        - The Effect: No single neuron can "monopolize" the learning. The knowledge becomes distributed across the whole network.
3. Properties and Characteristics
    - Randomness: Different neurons are "dropped" every time. A neuron that is off for Image A might be the main worker for Image B.
    - Redundancy: It creates multiple "internal paths" to the same answer. It's like having 5 different ways to get home; if one road is blocked (dropped), you still know the way.
    - Ensemble Effect: It’s like training thousands of slightly different, smaller networks and averaging their brains together.
4. How to Use It in PyTorch
    - In your nn.Sequential or class, you simply insert it between your linear layers.
5. Crucial Cautions (The "Buts")
    - Never use it during Testing!: This is the most important rule. If you turn off neurons while the model is trying to predict a real image, the answer will be garbage.
        - The Solution: Use model.eval(). This automatically tells PyTorch: "Wake up all the students; the test has started!"
    - Don't use it on the Input or Output layer: Usually, we only use Dropout on the Hidden Layers. You don't want to accidentally drop the raw pixels of the image or the final 10 categories!


1. `nn.Flatten()`
    - A standard nn.Linear layer is like a narrow pipe. It can only accept numbers coming in a single file line. You cannot push a square tile into a narrow pipe.
    - nn.Flatten() acts like a machine that breaks the tile into individual 1x1 pieces and lines them up.
    - Is it a layer? Yes, but it doesn't have any "weights" or "neurons." It doesn't learn anything; it just reshapes the data so the next layer can read it. In PyTorch, we call things "Layers" if they transform the data, even if they don't have parameters.

2. counting Parameters
    - This line of code tells you the total number of "knobs" your model can turn to learn.
    - model.parameters(): This is a list of all the weights and biases in your model.
    - p.numel(): As you noted in your first set of notes, this counts the total number of elements in a tensor.
    - Why is this number so big?
        - Even in your small model, you likely have over 100,000 parameters. Here is where they come from:
        - Layer 1: $784 \text{ inputs} \times 128 \text{ neurons} = 100,352$ weights (plus 128 biases).
        - Layer 2: $128 \times 64 = 8,192$ weights (plus 64 biases).
        - Output: $64 \times 10 = 640$ weights (plus 10 biases).
        

```python
class FashionMLP(nn.Module):
  def __init__(self):
    super().__init__()
    self.flatten = nn.Flatten()   # 28*28 -> 784
    self.hidden1 = nn.Linear(784,128)   # first hidden layer
    self.hidden2 = nn.Linear(128,64)    # second hidden layer
    self.output = nn.Linear(64,10)        # output layer: 10 classes
    self.relu = nn.ReLU()                 # activation for hidden layers
    self.dropout = nn.Dropout(0.2)        # regularization

  def forward(self, x):
    x = self.flatten(x)
    x = self.dropout(self.relu(self.hidden1(x)))
    x = self.dropout(self.relu(self.hidden2(x)))
    x = self.output(x)                            # Raw logits - NO softmax
    return x

# create model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = FashionMLP().to(device)
print(model)

# Count parameters
total_params = sum(p.numel() for p in model.parameters())
print(f"\nTotal parameters: {total_params:,}")
```

###  1.3.5 The training loop

```python
# Training Configuration
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

num_epochs = 10
print(f"Loss Function: CrossEntropyLoss (handles softmax internally)")
print(f"Optmizer: Adam (lr=0.001)")
print(f"Epochs: {num_epochs}")
```

```python
train_losses = []  # A notebook to keep track of how bad we were during training
test_losses = []   # A notebook for test scores
test_accs = []     # A notebook for accuracy %
start_time = time.time() # Start the stopwatch

for epoch in range(num_epochs):
  # ==== Training ====
  model.train()              # enable dropout
  running_loss = 0.0

  for batch_X, batch_y in train_loader:
    batch_X, batch_y = batch_X.to(device), batch_y.to(device)

    y_pred = model(batch_X)           # 1. Guess
    loss = loss_fn(y_pred, batch_y)   # 2. How wrong am I?
    optimizer.zero_grad()             # 3. Clear old memories (gradients)
    loss.backward()                   # 4. Do the calculus (backprop)
    optimizer.step()                  # 5. Move the weights (The Biker moves)

    running_loss += loss.item()

  avg_train_loss = running_loss / len(train_loader)
  train_losses.append(avg_train_loss)

  # === Evaluation
  model.eval()              # disable dropout
  test_loss = 0.0
  correct = 0
  total = 0

  with torch.inference_mode():
    for batch_X, batch_y in test_loader:
      batch_X, batch_y = batch_X.to(device), batch_y.to(device)
      y_pred = model(batch_X)
      test_loss += loss_fn(y_pred, batch_y).item()
      correct += (y_pred.argmax(dim=1) == batch_y).sum().item()
      total += batch_y.size(0)

  avg_test_loss = test_loss / len(test_loader)
  accuracy = correct / total
  test_losses.append(avg_test_loss)
  test_accs.append(accuracy)

  print(f"Epoch {epoch+1:2d}/{num_epochs} | " f"Train Loss: {avg_train_loss} | " f"Test Loss: {avg_test_loss} | " f"Test acc: {accuracy:.1%} | ")

elapsed = time.time() - start_time
print(f"Training complete in {elapsed:.1f}s")
print(f"Final test accuracy: {test_accs[-1]:.1%}")
```

The Setup (Lists & Timer)
- Purpose: To visualize the progress later (plotting a graph).
- If you don't write it: The model will still train, but you won't have any history to see if it was getting better or overfitting.

running_loss
- When you train in batches (e.g., 64 images at a time), PyTorch calculates the loss for just those 64 images.

Training: `model.train()`
- Purpose: This tells the model: "Wake up the Dropout!" * If you don't write it: If you have Dropout in your model, it might stay "off" from a previous session, and your model might start memorizing (overfitting).

The Mini-Batch Loop
- Purpose: Instead of 70,000 images at once, we take 64 images (batch_X) and their names (batch_y). We move them to the GPU (to(device)) so the math is fast.
- If you don't write it: You'll get a "Device Mismatch" error because your model is on the GPU but your data is still on the CPU.

The "Big Four" Math Steps
- If you don't write zero_grad(): The model will add the new errors to the old errors from the previous batch. The math will snowball and the model will fail to learn.

Evaluation: model.eval() & inference_mode()
- Purpose: This tells the model: "The test is starting! Disable Dropout and stop doing calculus." * If you don't write it: The model will keep "dropping" students during the test (randomly ignoring features), which makes your test score look much worse than it actually is.

Calculating Accuracy
- argmax(dim=1): Remember the 10 outputs (logits)? This finds the index of the highest number. If index 9 is the highest, it predicts "Ankle Boot."
- == batch_y: Compares the prediction to the real answer.
- If you don't write it: You'll only know the "Loss" (error), but you won't know if the model is 10% or 90% accurate.

```python
# Show predictions on test images
model.eval()
fig, axes = plt.subplots(2, 6, figsize=(15, 5.5))

with torch.inference_mode():
    for i, ax in enumerate(axes.flat):
        image, true_label = test_data[i + 10]  # skip first few for variety

        # Get prediction
        logits = model(image.unsqueeze(0).to(device))
        probs = torch.softmax(logits, dim=1)   # NOW we apply softmax for probabilities
        pred_label = probs.argmax(dim=1).item()
        confidence = probs.max().item()

        # Display
        ax.imshow(image.squeeze(), cmap='gray')
        color = 'green' if pred_label == true_label else 'red'
        ax.set_title(f"{class_names[pred_label]}\n({confidence:.0%})",
                     fontsize=10, color=color, fontweight='bold')
        ax.axis('off')

plt.suptitle('Model Predictions (green = correct, red = wrong)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
